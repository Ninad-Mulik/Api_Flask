```python
import matplotlib.pyplot as plt

acc_data = {
    "TF CNN Model": [0.9, 0.92, 0.94, 0.96, 0.98],  # Increasing accuracy values for TensorFlow
    "TF lite CNN Model": [0.85, 0.87, 0.89, 0.91, 0.93],  # Increasing accuracy values for TensorFlow Lite
    "VGG16": [0.8, 0.82, 0.84, 0.86, 0.88],  # Increasing accuracy values for VGG16
}

colors = ["#1f77b4", "#ff7f0e", "#2ca02c"]  # Set custom colors

fig, ax = plt.subplots(figsize=(10, 6))  # Adjust the figure size

bar_width = 0.2
num_models = len(acc_data)
index = range(len(acc_data["TF CNN Model"]))  # Assuming all models have the same number of data points

for i, model in enumerate(acc_data):
    ax.bar(
        [pos + i * bar_width for pos in index],  # X-axis positions
        acc_data[model],  # Heights
        bar_width,
        label=model,
        color=colors[i % len(colors)],
        edgecolor='black',  # Add black border to bars
        linewidth=1.5,  # Increase border width
    )

ax.set_xlabel("Training Percentage", fontsize=14)  # Increase font size
ax.set_ylabel("Accuracy", fontsize=15)  # Increase font size
ax.tick_params(axis='both', which='major', labelsize=12)  # Increase tick label font size

# Manually set the position of x-axis ticks and adjust labels
ax.set_xticks([pos + (num_models * bar_width) / 2 for pos in index])
ax.set_xticklabels(["40%", "50%", "60%", "70%", "80%"], ha='center')

ax.set_title("Model Accuracy Per Training Percentage", fontsize=16)  # Increase title font size
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), shadow=True, ncol=3, fontsize=12)  # Adjust legend font size and position

plt.grid(axis='y', linestyle='', alpha=0.7)  # Add horizontal grid lines
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
